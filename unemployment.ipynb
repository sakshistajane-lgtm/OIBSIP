```python
# Import Libraries
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
data= pd.read_csv("Unemployment.csv")

# Rename columns(if needed)
data.columns= data.columns.str.strip()

#plot Unemployment Rate by Date
plt.figure(figsize=(10,5))

sns.lineplot(
    x="Date",
    y="Estimated Unemployment Rate (%)",
    data=data )

plt.title("Unemployment Rate over time")
plt.xticks(rotation=45)
plt.show()

# state_wise Average Unemployment 
data["Estimated Unemployment Rate (%)"]= pd.to_numeric(
    data["Estimated Unemployment Rate (%)"],
    errors="coerce"
)
state_avg= data.groupby("Region")["Estimated Unemployment Rate (%)"].mean()
print(state_avg)

#Bar graph
plt.figure(figsize=(10,6))

state_avg.sort_values().plot(kind= "barh")

plt.title("Average Unemployment Rate by State")
plt.xlabel("Unemployment Rate (%)")
plt.show()
```
